```python
def get_billboard_chart(date):
    URL = f"https://www.billboard.com/charts/hot-100/{date}/"
    response = requests.get(URL)
    soup = BS(response.text)

    ranks = soup.findAll('span', attrs={'class' : 'u-line-height-33px'})
    rank_this_week = [x.get_text(strip=True) for x in ranks]

    artists = soup.findAll('span', attrs={'class' : 'u-line-height-21px'})
    artist_this_week = [x.get_text(strip=True) for x in artists]

    songs = soup.findAll('h3', attrs={'class' : 'u-line-height-22px'})
    song_this_week = [x.get_text(strip=True) for x in songs]

    other_ranks_and_weeks = soup.findAll('span', attrs={'class' : 'u-min-width-30px'})
    weeks_and_other_ranks = [x.get_text(strip=True) for x in other_ranks_and_weeks]

    last_week = []
    if weeks_and_other_ranks:
        # Add the first element's text
        last_week.append(weeks_and_other_ranks[0])
    
        # Iterate from the third element (index 2) and take every third one
        for i in range(6, len(weeks_and_other_ranks), 6):
            last_week.append(weeks_and_other_ranks[i])
    
    
    peak = []
    if weeks_and_other_ranks:
        # Add the first element's text
        peak.append(weeks_and_other_ranks[1])
    
        # Iterate from the third element (index 2) and take every third one
        for i in range(7, len(weeks_and_other_ranks), 6):
            peak.append(weeks_and_other_ranks[i])
    
    weeks = []
    if weeks_and_other_ranks:
        # Add the first element's text
        weeks.append(weeks_and_other_ranks[2])
    
        # Iterate from the third element (index 2) and take every third one
        for i in range(8, len(weeks_and_other_ranks), 6):
            weeks.append(weeks_and_other_ranks[i])

    data = {
        'This Week' : rank_this_week,
        'Artist' : artist_this_week,
        'Song' : song_this_week,
        'Last Week' : last_week,
        'Peak' : peak,
        'Weeks on Chart' : weeks 
    }

    df = pd.DataFrame(data)

    return df
```

```python
get_billboard_chart('2025-10-04')
```

```python
from datetime import date, timedelta

def get_past_saturdays(num_saturdays):
    """
    Generates a list of date strings for the past 'num_saturdays' Saturdays.
    """
    today = date.today()
    saturdays = []

    # Find the most recent Saturday (could be today if today is Saturday)
    # weekday() returns 0 for Monday, 6 for Sunday. Saturday is 5.
    days_since_saturday = (today.weekday() - 5 + 7) % 7
    last_saturday = today - timedelta(days=days_since_saturday)

    for i in range(num_saturdays):
        current_saturday = last_saturday - timedelta(weeks=i)
        saturdays.append(current_saturday.isoformat()) # Format as YYYY-MM-DD
    return saturdays

# Get the past 10 Saturdays
past_10_saturdays = get_past_saturdays(10)

# Loop through the dates and pass them to the function
for saturday_date_str in past_10_saturdays:
    get_billboard_chart(saturday_date_str)
```
