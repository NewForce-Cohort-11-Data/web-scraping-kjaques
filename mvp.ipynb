```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import re
```

```python
URL = 'https://realpython.github.io/fake-jobs/'

response = requests.get(URL)
```

```python
type(response)
```

```python
response.status_code
```

```python
soup = BS(response.text)
```

```python
soup.find('h2').text
```

```python
type(soup.find('h2'))
```

```python
titles = soup.findAll('h2')
```

```python
first_title = titles[0]
first_title
```

```python
job_titles = [x.get_text() for x in titles]
```

```python
soup.find('h3').text
```

```python
companies = soup.findAll('h3')
```

```python
job_companies = [x.get_text() for x in companies]
```

```python
locations = soup.findAll('p', attrs={'class' : 'location'})
```

```python
job_locations = [x.get_text(strip=True, separator=',') for x in locations]
```

```python
posting_date = soup.findAll('time')
```

```python
job_date = [x.get_text() for x in posting_date]
```

```python
data = {'Job Title': job_titles,
        'Company': job_companies,
        'Location': job_locations,
        'Date Posted' : job_date}
jobs = pd.DataFrame(data)
jobs
```

```python
urls = soup.findAll('a', string=re.compile('Apply'))
job_urls = [x.get('href') for x in urls]
```

```python
job_urls
```

```python
urls_alt = jobs['Job Title'].str.lower()
```

```python
urls_alt = urls_alt.str.replace(r'[\W_]','-', regex=True)
```

```python
urls_alt = urls_alt.str.replace('--','-')
```

```python
urls_alt = urls_alt.str.replace(r'-$','', regex=True)
```

```python
urls_alt.tolist()
```

```python
url_prefix = 'https://realpython.github.io/fake-jobs/jobs/'
```

```python
cleaned_urls = [url_prefix + str(val) + "-" + str(idx) + '.html' for idx, val in enumerate(urls_alt)]
```

```python
def scrape_descriptions(df, url_column):
    results = []
    
    for url in df[url_column]:
        try:
            response = requests.get(url, timeout=10)
            response.raise_for_status()  # raises error for bad status codes
            
            soup = BS(response.text, "html.parser")
            # Find first <p> with no class or id
            p_tag = soup.find("p", class_=False, id=False)
            
            if p_tag and p_tag.text.strip():
                results.append(p_tag.text.strip())
            else:
                results.append(None)  # No matching <p>
        
        except Exception as e:
            print(f"Error scraping {url}: {e}")
            results.append(None)
    
    return results
```

```python
scraped_descriptions = scrape_descriptions(jobs, 'URL')

# Add back to DataFrame if you like:
jobs['Description'] = scraped_descriptions
```
